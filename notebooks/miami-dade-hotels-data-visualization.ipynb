```python
import folium
import pandas as pd
```

```python
htl = pd.read_csv ('/workspaces/codespaces-jupyter/data/Hotel_Motel_Inn.csv')
htl.head(558)
```

```python
mia_map = folium. Map(
    location=[25.761681, -80.191788],
    zoom_start=12
)
for _, h in htl.iterrows():
    folium. Marker(
        location=[h['LAT'], h['LON']],
        popup=h['CAPACITY'],
        tooltip=h['NAME'],
        icon=folium.Icon(color="green", icon="home") 
    ).add_to (mia_map)

mia_map
```
